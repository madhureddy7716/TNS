#**Matplotlib**

1. Line Plot

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [1000, 1500, 1800, 2200, 3000]

plt.plot(months, sales, marker='o')
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

2. Bar Chart


```python
import matplotlib.pyplot as plt

students = ["A", "B", "C", "D", "E"]
marks = [80, 92, 75, 88, 95]

plt.bar(students, marks)

plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

3. Histogram

```python
import matplotlib.pyplot as plt

ages = [18,19,20,21,22,22,23,24,25,25,26,27,28,28,29]

plt.hist(ages, bins=5)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

4. Scatter Plot

```python
import matplotlib.pyplot as plt

hours = [1,2,3,4,5,6,7]
marks = [30,40,50,60,70,85,95]

plt.scatter(hours, marks)

plt.title("Study Hours vs Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")

plt.show()
```

5. Pie Chart


```python
import matplotlib.pyplot as plt

subjects = ["Python", "Java", "C++", "JavaScript"]
students = [40, 25, 20, 15]

plt.pie(students,
        labels=subjects,
        autopct="%1.1f%%")

plt.title("Programming Language Popularity")

plt.show()
```

6. Multiple Graphs

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]

plt.subplot(1,2,1)
plt.plot(x,[1,4,9,16,25])
plt.title("Square")

plt.subplot(1,2,2)
plt.plot(x,[1,8,27,64,125])
plt.title("Cube")

plt.show()
```

plt.title() – Title

```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [10,20,30]

plt.plot(x,y)
plt.title("Student Marks")
plt.show()
```

plt.xlabel() – X-Axis Label

```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [10,20,30]

plt.plot(x,y)
plt.xlabel("Students")
plt.show()
```

plt.ylabel() – Y-Axis Label

```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [10,20,30]

plt.plot(x,y)
plt.ylabel("Marks")
plt.show()
```

plt.grid() – Grid

```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [10,20,30]

plt.plot(x,y)
plt.grid(True)
plt.show()
```

plt.figure() – Figure Size

```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [10,20,30]

plt.figure(figsize=(8,4))
plt.plot(x,y)
plt.show()
```

Marker

```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [10,20,30]

plt.plot(x,y,marker='o')
plt.show()
```

Color

```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [10,20,30]

plt.plot(x,y,color='red')
plt.show()
```
