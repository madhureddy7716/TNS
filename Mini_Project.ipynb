Create a Student Report Card Program using:

1.List → store student names

2.Set → store subjects

3.Tuple → store student information (name, age)

4.Dictionary → store subject-wise marks

Requirements:

1.Ask the user to enter:

*Student name

*Age

*Number of subjects

2.For each subject:

*Enter subject name

*Enter marks

3.Store:

*Name in a list

*Subjects in a set

*Student info in a tuple

*Marks in a dictionary

4.Print a report card showing:

*Name

*Age

*Subjects

*Marks

*Total marks

*Average marks   make with answer

```python
# Student Report Card Program

# List to store student names
student_names = []

# Set to store subjects
subjects = set()

# Dictionary to store subject-wise marks
marks_dict = {}

# Input student details
name = input("Enter Student Name: ")
age = int(input("Enter Age: "))
num_subjects = int(input("Enter Number of Subjects: "))

# Store name in list
student_names.append(name)

# Store student info in tuple
student_info = (name, age)

# Input subjects and marks
for i in range(num_subjects):
    subject = input(f"Enter Subject {i+1} Name: ")
    marks = int(input(f"Enter Marks in {subject}: "))

    subjects.add(subject)
    marks_dict[subject] = marks

# Calculate total and average
total_marks = sum(marks_dict.values())
average_marks = total_marks / num_subjects

# Report Card
print("\n===== STUDENT REPORT CARD =====")
print("Name:", student_info[0])
print("Age:", student_info[1])
print("Subjects:", subjects)

print("\nSubject-wise Marks:")
for subject, marks in marks_dict.items():
    print(subject, ":", marks)

print("\nTotal Marks:", total_marks)
print("Average Marks:", average_marks)
```

# **Pandas Mini Project**

1. Student Management System

```python
import pandas as pd

data = {
    "Name": ["Madhu", "Ravi", "Priya", "Kiran", "Anu"],
    "Maths": [85, 90, 78, 88, 95],
    "Science": [80, 85, 75, 92, 90],
    "English": [70, 88, 82, 80, 95]
}

df = pd.DataFrame(data)

df["Total"] = df["Maths"] + df["Science"] + df["English"]
df["Average"] = df["Total"] / 3

df["Grade"] = df["Average"].apply(
    lambda x: "A" if x >= 90 else
              "B" if x >= 80 else
              "C"
)

print("Student Report")
print(df)

print("\nTopper")
print(df.loc[df["Total"].idxmax()])

print("\nLowest Scorer")
print(df.loc[df["Total"].idxmin()])
```

Employee Salary Management

```python
import pandas as pd

data = {
    "Employee": ["A", "B", "C", "D", "E"],
    "Department": ["IT", "HR", "IT", "Sales", "HR"],
    "Salary": [50000, 40000, 60000, 45000, 55000]
}

df = pd.DataFrame(data)

df["Bonus"] = df["Salary"] * 0.10

print("Employee Data")
print(df)

print("\nAverage Salary")
print(df["Salary"].mean())

print("\nHighest Salary")
print(df.loc[df["Salary"].idxmax()])

print("\nDepartment Wise Salary")
print(df.groupby("Department")["Salary"].mean())
```

Cricket Statistics System

```python
import pandas as pd

data = {
    "Player": ["Virat", "Rohit", "Gill", "Rahul", "Pant"],
    "Runs": [120, 80, 95, 45, 110],
    "Matches": [3, 3, 3, 3, 3]
}

df = pd.DataFrame(data)

df["Average"] = df["Runs"] / df["Matches"]

print("Cricket Statistics")
print(df)

print("\nHighest Run Scorer")
print(df.loc[df["Runs"].idxmax()])

print("\nLowest Run Scorer")
print(df.loc[df["Runs"].idxmin()])

print("\nPlayers with Runs > 100")
print(df[df["Runs"] > 100])

print("\nSorted by Runs")
print(df.sort_values(by="Runs", ascending=False))
```

Sales Analysis System

```python
import pandas as pd

data = {
    "Product": ["Laptop", "Mobile", "Laptop", "TV", "Mobile"],
    "Quantity": [2, 5, 3, 1, 4],
    "Price": [50000, 20000, 50000, 30000, 20000]
}

df = pd.DataFrame(data)

df["Sales"] = df["Quantity"] * df["Price"]

print("Sales Report")
print(df)

print("\nTotal Revenue")
print(df["Sales"].sum())

print("\nProduct Wise Revenue")
print(df.groupby("Product")["Sales"].sum())

print("\nHighest Sale")
print(df.loc[df["Sales"].idxmax()])
```


# **Task-3 :Hospital Patient Data Analysis**

```python
# ==========================================
# Hospital Patient Data Analysis Project
# ==========================================

import pandas as pd
import numpy as np

# ------------------------------------------
# Dataset
# ------------------------------------------

data = {
    "Patient_ID": [101, 102, 103, 104, 105, 106, 106],
    "Name": ["Madhu", "Suresh", "Lakshmi", "Ramesh", "Kavya", "Vijay", "Vijay"],
    "Age": [45, 30, 55, np.nan, 40, 60, 60],
    "Disease": ["Diabetes", "Cancer", "Heart Disease", "Diabetes", "Cancer", "Asthma", "Asthma"],
    "Treatment_Cost": [50000, 120000, 80000, 60000, np.nan, 30000, 30000],
    "Days_Admitted": [10, 20, 15, 12, 18, 5, 5]
}

# Create DataFrame
df = pd.DataFrame(data)

# ------------------------------------------
# 1. Load and Inspect Dataset
# ------------------------------------------

print("=" * 60)
print("HOSPITAL PATIENT DATASET")
print("=" * 60)

print("\nDataset:")
print(df)

print("\nDataset Information:")
print(df.info())

print("\nFirst 5 Records:")
print(df.head())

print("\nMissing Values:")
print(df.isnull().sum())

# ------------------------------------------
# 2. Handle Missing Values
# ------------------------------------------

df["Age"] = df["Age"].fillna(df["Age"].mean())

df["Treatment_Cost"] = df["Treatment_Cost"].fillna(
    df["Treatment_Cost"].mean()
)

print("\n" + "=" * 60)
print("AFTER HANDLING MISSING VALUES")
print("=" * 60)

print(df)

# ------------------------------------------
# 3. Remove Duplicate Records
# ------------------------------------------

df = df.drop_duplicates()

print("\n" + "=" * 60)
print("AFTER REMOVING DUPLICATES")
print("=" * 60)

print(df)

# ------------------------------------------
# 4. Calculate Cost Per Day
# ------------------------------------------

df["Cost_Per_Day"] = (
    df["Treatment_Cost"] / df["Days_Admitted"]
)

print("\n" + "=" * 60)
print("COST PER DAY")
print("=" * 60)

print(df[["Name", "Cost_Per_Day"]])

# ------------------------------------------
# 5. Find Patients with Treatment Costs
#    Above Average
# ------------------------------------------

average_cost = df["Treatment_Cost"].mean()

above_average = df[
    df["Treatment_Cost"] > average_cost
]

print("\n" + "=" * 60)
print("PATIENTS WITH ABOVE AVERAGE COST")
print("=" * 60)

print(above_average[["Name", "Disease", "Treatment_Cost"]])

# ------------------------------------------
# 6. Sort by Treatment Cost
# ------------------------------------------

sorted_df = df.sort_values(
    by="Treatment_Cost",
    ascending=False
)

print("\n" + "=" * 60)
print("SORTED BY TREATMENT COST")
print("=" * 60)

print(sorted_df[["Name", "Treatment_Cost"]])

# ------------------------------------------
# 7. Disease-wise Average Treatment Cost
# ------------------------------------------

disease_average = df.groupby(
    "Disease"
)["Treatment_Cost"].mean()

print("\n" + "=" * 60)
print("DISEASE-WISE AVERAGE TREATMENT COST")
print("=" * 60)

print(disease_average)

# ------------------------------------------
# 8. Apply NumPy Statistical Functions
# ------------------------------------------

costs = df["Treatment_Cost"].to_numpy()

print("\n" + "=" * 60)
print("NUMPY STATISTICS")
print("=" * 60)

print("Mean Cost              :", np.mean(costs))
print("Median Cost            :", np.median(costs))
print("Maximum Cost           :", np.max(costs))
print("Minimum Cost           :", np.min(costs))
print("Standard Deviation     :", np.std(costs))

# ------------------------------------------
# 9. Find Most Expensive Treatment
# ------------------------------------------

most_expensive = df.loc[
    df["Treatment_Cost"].idxmax()
]

print("\n" + "=" * 60)
print("MOST EXPENSIVE TREATMENT")
print("=" * 60)

print("Patient Name     :", most_expensive["Name"])
print("Disease          :", most_expensive["Disease"])
print("Treatment Cost   :", most_expensive["Treatment_Cost"])
print("Days Admitted    :", most_expensive["Days_Admitted"])

# ------------------------------------------
# 10. Generate Hospital Report
# ------------------------------------------

print("\n" + "=" * 60)
print("HOSPITAL REPORT")
print("=" * 60)

print("Total Patients           :", len(df))
print("Average Age              :", round(df["Age"].mean(), 2))
print("Average Treatment Cost   :", round(df["Treatment_Cost"].mean(), 2))
print("Highest Treatment Cost   :", df["Treatment_Cost"].max())
print("Lowest Treatment Cost    :", df["Treatment_Cost"].min())

print("\nDisease-wise Average Cost:")
print(disease_average)

print("\nMost Expensive Treatment:")
print(most_expensive[["Name", "Disease", "Treatment_Cost"]])

print("\nReport Generated Successfully!")
print("=" * 60)
```
