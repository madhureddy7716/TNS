## Pandas Practice Set

This single program covers:

```python
import pandas as pd
import numpy as np

# =====================================
# 1. SERIES
# =====================================

print("===== SERIES =====")

s = pd.Series([10, 20, 30, 40, 50])

print(s)

print("Mean:", s.mean())
print("Std:", s.std())
print("Min:", s.min())
print("Max:", s.max())
print("Sum:", s.sum())

# =====================================
# 2. MAP AND APPLY
# =====================================

print("\n===== MAP =====")

double = s.map(lambda x: x * 2)
print(double)

print("\n===== APPLY =====")

square = s.apply(lambda x: x ** 2)
print(square)

# =====================================
# 3. SORTING
# =====================================

print("\n===== SORTING =====")

s1 = pd.Series([45, 12, 78, 2, 90, 35])

print(s1.sort_values())
print(s1.sort_values(ascending=False))

# =====================================
# 4. DROP
# =====================================

print("\n===== DROP =====")

print(s.drop(2))

# =====================================
# 5. MISSING VALUES
# =====================================

print("\n===== MISSING VALUES =====")

s2 = pd.Series([10, 20, np.nan, 40, np.nan])

print(s2)

print("isnull")
print(s2.isnull())

print("notnull")
print(s2.notnull())

print("fillna")
print(s2.fillna(0))

print("dropna")
print(s2.dropna())

# =====================================
# 6. INDEXING
# =====================================

print("\n===== INDEXING =====")

print(s.iloc[2])

print(s.iloc[1:4])

print(s.loc[1:3])

# =====================================
# 7. FILTERING
# =====================================

print("\n===== FILTERING =====")

print(s[s > 25])

# =====================================
# 8. AGGREGATION
# =====================================

print("\n===== AGGREGATION =====")

print(s.sum())

print(s.cumsum())

print(s.aggregate(['sum', 'mean', 'std']))

# =====================================
# 9. DATAFRAME
# =====================================

print("\n===== DATAFRAME =====")

data = {
    "Name": ["Madhu", "Ravi", "Priya", "Kiran", "Anu"],
    "Age": [22, 21, 23, 20, 24],
    "Marks": [85, 90, 78, 88, 95]
}

df = pd.DataFrame(data)

print(df)

# =====================================
# 10. HEAD AND TAIL
# =====================================

print("\n===== HEAD =====")
print(df.head())

print("\n===== TAIL =====")
print(df.tail())

# =====================================
# 11. SHAPE AND COLUMNS
# =====================================

print("\n===== SHAPE =====")
print(df.shape)

print("\n===== COLUMNS =====")
print(df.columns)

# =====================================
# 12. DESCRIBE
# =====================================

print("\n===== DESCRIBE =====")
print(df.describe())

# =====================================
# 13. FILTER DATAFRAME
# =====================================

print("\n===== FILTER =====")

print(df[df["Marks"] > 85])

# =====================================
# 14. ADD NEW COLUMN
# =====================================

print("\n===== NEW COLUMN =====")

df["Result"] = df["Marks"].apply(
    lambda x: "Pass" if x >= 40 else "Fail"
)

print(df)

# =====================================
# 15. SORT DATAFRAME
# =====================================

print("\n===== SORT MARKS =====")

print(df.sort_values(by="Marks"))

print(df.sort_values(by="Marks", ascending=False))

# =====================================
# 16. GROUPBY
# =====================================

print("\n===== GROUPBY =====")

emp = {
    "Department": ["IT", "IT", "HR", "HR", "Sales", "Sales"],
    "Salary": [50000, 60000, 45000, 55000, 40000, 70000]
}

emp_df = pd.DataFrame(emp)

print(emp_df.groupby("Department")["Salary"].mean())

print(emp_df.groupby("Department")["Salary"].max())

print(emp_df.groupby("Department")["Salary"].sum())

# =====================================
# 17. MERGE
# =====================================

print("\n===== MERGE =====")

df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Madhu", "Ravi", "Priya"]
})

df2 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Branch": ["CSE", "ECE", "AIML"]
})

merged = pd.merge(df1, df2, on="ID")

print(merged)

# =====================================
# 18. SAVE CSV
# =====================================

df.to_csv("students.csv", index=False)

print("\nCSV Saved Successfully")

# =====================================
# 19. READ CSV
# =====================================

new_df = pd.read_csv("students.csv")

print(new_df)

# =====================================
# 20. INFO
# =====================================

print("\n===== INFO =====")

print(new_df.info())
```
