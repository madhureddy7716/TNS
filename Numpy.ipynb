1. Array Creation


```python
import numpy as np

arr1 = np.array([10,20,30,40,50])
print(arr1)

arr2 = np.array([[1,2,3],
                 [4,5,6]])
print(arr2)

zeros = np.zeros((3,4))
print(zeros)

ones = np.ones((4,4))
print(ones)

identity = np.eye(5)
print(identity)

arange_arr = np.arange(1,21)
print(arange_arr)

linspace_arr = np.linspace(0,100,10)
print(linspace_arr)
```

2. Array Properties


```python
arr = np.array([[1,2,3],
                [4,5,6]])

print("Shape:", arr.shape)
print("Dimensions:", arr.ndim)
print("Size:", arr.size)
print("Datatype:", arr.dtype)

float_arr = arr.astype(float)
print(float_arr)
```

3. Reshape and Transpose

```python
arr = np.arange(1,13)

print(arr)

print(arr.reshape(3,4))
print(arr.reshape(2,6))
print(arr.reshape(4,3))

matrix = arr.reshape(3,4)
print(matrix.T)
```

4. Mathematical Operations


```python
arr = np.array([10,20,30,40,50])

print(arr + 5)
print(arr - 10)
print(arr * 2)
print(arr / 5)

print(arr ** 2)
print(arr ** 3)
```

5. Aggregate Functions

```python
arr = np.array([12,25,18,40,50])

print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Std:", np.std(arr))
print("Variance:", np.var(arr))

print("Min:", np.min(arr))
print("Max:", np.max(arr))
```

6. Indexing and Slicing

```python
arr = np.array([10,20,30,40,50,60,70])

print(arr[0])
print(arr[-1])

print(arr[:3])

print(arr[2:6])

print(arr[::2])
```

7. Random Numbers

```python
rand_arr = np.random.rand(2,3)
print(rand_arr)

rand_int = np.random.randint(1,11,(3,3))
print(rand_int)

otp = np.random.randint(100000,999999)
print("OTP:", otp)

random_numbers = np.random.rand(5)
print(random_numbers)
```

8. Sorting

```python
arr = np.array([9,2,7,1,5,8])

print(np.sort(arr))

print(np.sort(arr)[::-1])

print(np.argmax(arr))

print(np.argmin(arr))
```

9. Matrix Operations

```python
A = np.array([[1,2],
              [3,4]])

B = np.array([[5,6],
              [7,8]])

print(A + B)

print(A - B)

print(np.dot(A,B))

print(A * B)

print(A.T)
```
